```python
import numpy as np
import pandas as pd
from scipy.spatial.distance import cosine
from scipy.stats.stats import pearsonr
```

```python
data = pd.read_pickle('../clean-data/dataf.pickle')
```

```python
data1 = data.fillna(0)
nmat = data1.as_matrix()
```

```python
def cos(a, b):
    a = np.array(a)
    b = np.array(b)
    return 1 - cosine(a,b)
```

```python
def euc(a, b):
    a = np.array(a)
    b = np.array(b)
    return 1.0/ (1 + np.linalg.norm(a-b))
    
```

```python
def pr(a,b):
    a = np.array(a)
    b = np.array(b)   
    return pearsonr(a,b)[0]
```

```python
def sim(matrix, a, b, t = 'pr'):
    aval = matrix[a].nonzero()
    bval = matrix[b].nonzero()
    shared = np.intersect1d(aval, bval)
    print(shared)
    if len(shared) == 0:
        return 0
    a = np.take(matrix[a], shared)
    print(a)
    b = np.take(matrix[b], shared)
    print(b)
    if t =='euc':
        return euc(a,b)
    elif t == 'cos':
        return cos(a,b)
    else:
        return pr(a,b) 
```

```python
def user_user(a,n):
    score_list={}
    for b in udict.keys():
        if(b!=a):#Compute Pearson Correlation Value for Every Other User in the List
            score_list[b]=sim(nmat, a, b)       
    sorteddict = sorted(score_list.values())
    return sorteddict[-n:]
```
